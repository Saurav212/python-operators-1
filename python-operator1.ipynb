```python
# Python operators - Operators are used to perform operations on variables and values.

# Python divides the operators in the following groups:

# Arithmetic operators
# Assignment operators
# Comparison operators
# Logical operators
# dentity operators
# Membership operators
# Bitwise operators

# Arithmetic Operators :- Arithmetic operators are used with numeric values to perform common mathematical operations:

#  Operator    Name          Example
#     +     Addition          x + y
#     -     Subtraction       x - y
#     *     Multiplication    x * y
#     /     Division          x / y
#     %    Modulus            x % y
#    **    Exponentiation     x ** y
#    //    Floor division     x // y

```

```python
# Addition Operator(+) - It adds twumbers and concatenates strings.

x, y = 5, 10
print("x + y =", x+y)
```

```python
# Concatenation

str1 = "Hi"
str2 = "Saurav"
print(str1 + " " + str2)
```

```python
# Subtraction Operator(-) - It subtract two numbers.

x,y = 30, 10
print("x - y =", x-y)
```

```python
# Multiplication Operator(*) - It multiplies two numbers.

x, y = 5, 15 
print ("x*y = ", x*y)
```

```python
# Division Operator(/) :- It divides two numbers.

x, y = 5, 15 
print ("y/x = ", y/x)
```

```python
# Modulus Operator(%) :- It gives the remainder when two numbers are divided.

x, y = 50, 15 
print ("x % y = ", x%y)
```

```python
# Exponentiation Operator(**) :- It gives the exponent.

x, y = 5, 5 
print ("x ** y = ", x**y)
```

```python
# Floor Division Operator(//) :- Floor division returns the quotient(answer or result of division) in which the digits after the decimal point are removed. But if one of the operands(dividend and divisor) is negative, then the result is floored, i.e., rounded away from zero(means, towards the negative of infinity).

# example -1
x, y = 559, 5 
print ("x // y = ", x // y)
```

```python
# example -2
x, y = 560, 5 
print ("x // y = ", x // y)
```

```python
# example -3
x, y = -559, 5 
print ("x // y = ", x // y)
```

```python
# example -4

x, y = 559, -5 
print ("x // y = ", x // y)
```

```python
# example -5

x, y = 0, 5 
print ("x // y = ", x // y)
```
